### Фильтрация данных:

```python
# фильтрация во многом похоже на трактовку условных операторов в SQL и  python
df[df['Age'] < 35].head(5)
```

```python
# что бы выбрать конкретные столбы по условию напишем так: 
df[['PassengerId', 'Name', 'Lname', 'Sex', 'Age']]\
    [df['Age']>35].head(5)
```

```python
# кроме привычной фильтрации можно просто преверить по всем столбам логическое значение по условие(правдиво ли оно или нет)
df['Age'] > 35
```

```python
# что бы не считать руками проверку логического условия выше - можно воспользоваться методом: 
df[df['Age'] > 35].shape

# важно понимать, что при подсчете таким образом - метод shape ожидает не получить не проверку условия - а само условие
# поэтому само написание будет немного разным
```

```python
# давайте отфильтруем некоторые значения с помощь `.isin`

df[['PassengerId', 'Name', 'Lname', 'Sex', 'Age']]\
    [df['Age'].isin([25,65,43])].head(5)
```

```python
# а теперь разберем код выше на переменные
# зададим конкретные поля в переменной
fields = df[['PassengerId', 'Name', 'Lname', 'Sex', 'Age']]

# отдадим переменную методу .isin и внутри аргумента .isin передадим список интересующих значений
fields[df['Age'].isin([25,65,43])].head()
```

```python
# код выше можно написать и с применением логического оператора так: 
fields = df[['PassengerId', 'Name', 'Lname', 'Sex', 'Age']]
fields[(df['Age'] == 25) | (df['Age'] == 65) | (df['Age'] == 43)].head(5)

# но .isin выглядит более понятным и лаконичным - и позволит не запутаться в скобках
```

```python
# допустим каких-то данных нет в дф, что плохо для статистики, поэтому мы хотим избавится от таких данных
# для этого есть специальный метод notna()

# Условная функция возвращает для Trueкаждой строки значения, не являющиеся Null значением.
#  Таким образом, это можно комбинировать со скобками выбора [] для фильтрации таблицы данных.

# т.е. -> fields.notna().head(5) -> такой строкой мы снова вернем проверку соотвествия 
# условию логического выражения, но не сами данные

fields = df[['PassengerId', 'Name', 'Lname', 'Sex', 'Age']]
# и зададим условие методом .notna()
fields[df['Age'].notna()].head(5)
```

```python
# есть еще один способ отфильтровать по условию и вывести конкретные столбы 
df.loc[df['Age']>35, 'Age'].head(5)

# такой способ лаконичен, но не позволит вывести несколько столбцов разом
```

```python
# если известны данные и индексы(строки) где они лежат -> их можно достать по срезу строк + срезу столбцов
df.iloc[1:10, 0:6]

# это может заменить длинные указания имен столбцов, 
# но только в том случае если полностью известна структура рассматриваемого дф
```

```python
# или так
# Опять же, подмножество как строк, так и столбцов создается за один раз, и 
# простого использования скобок выбора []уже недостаточно. Если вас особенно интересуют 
# определенные строки и/или столбцы в зависимости от их положения в таблице, используйте iloc 
# оператор перед скобками выбора [].
param = df.iloc[1:10, 0:5]
param.head(5)
```

```python
# кроме вывода - можно также и манипулировать данными (менять)
df.iloc[2:3, 3] = 'Futrelle'
df.head()
```

### Важно
- При выборе подмножеств данных используются квадратные скобки [ ].
- Внутри этих скобок вы можете использовать одну метку столбца/строки, список меток столбца/строки, часть меток, условное выражение или двоеточие.
- Выберите определенные строки и/или столбцы с помощью loc при использовании имен строк и столбцов
- Выберите определенные строки и/или столбцы, используя iloc при использовании позиций в таблице
- Вы можете присвоить новые значения выбору на основе loc/ iloc.
