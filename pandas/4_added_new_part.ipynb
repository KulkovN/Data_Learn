```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/air_quality_no2.csv')
df.head()
```

```python
# проверим сотношение значений по NO2 нескольких городов и пропишем их в отдельный столб
df ['ratio_paris_london'] = (
    df['station_paris'] / df['station_london'])
df.head()
```

```python
# для переименования столбов: 
df_renamed_ratio_p_l = df.rename(
    columns = {
        "ratio_paris_london" : "p_l-stations_ratio",
        }
    )
df_renamed_ratio_p_l.head()

# как видим, создается реплика дф, и столбы уже имеют вновь-заданное имя
# новые значения при этом указываются в дикте, а ключи == старые имена, значения == новые значение
```
